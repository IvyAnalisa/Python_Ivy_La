2.2 Finding and dropping NA values >Lets now check if there are any NA values in any of our columns>

```python
import pandas as pd

# Load the wine quality dataset
df = pd.read_csv("winequality-red.csv")

def count_nan():
    # Total number of samples (rows) in the dataset
    total_samples = len(df)

    # For each column, calculate the percentage of missing values
    for col in df.columns:
        missing_count = df[col].isna().sum()  # Number of missing values in the column
        missing_percentage = (missing_count / total_samples) * 100  # Percentage of missing values

        # Display the result for the current column with the exact format requested
        print(f"The column {col} is missing {missing_percentage:.4f}% of data!")
        
# Call the function to display missing value percentages
count_nan()
```

# Next up is a bit of a tricky part; We will now identify every sample that isna but instead of summarizing we will identify the samples such that we later on can remove them. Once again use isna() but this time say all() instead of sum()! You might also need to specify the axis

```python
import pandas as pd

# Load the wine quality dataset
df = pd.read_csv("winequality-red.csv")

# 1. Extract all the samples that meet the criteria of isna() in any column
nan_samples = df[df.isna().any(axis=1)]  # Filters rows with any NaN values

# 2. Get the indexes for these rows
nan_indexes = nan_samples.index  # Extract the index of rows with NaN values

# 3. Drop these rows from the DataFrame
df = df.drop(nan_indexes)  # Drop the rows with NaN values from the DataFrame

# Define count_nan function to print the percentage of missing values
def count_nan():
    # Total number of samples (rows) in the dataset
    total_samples = len(df)

    # For each column, calculate the percentage of missing values
    for col in df.columns:
        missing_count = df[col].isna().sum()  # Number of missing values in the column
        missing_percentage = (missing_count / total_samples) * 100  # Percentage of missing values

        # Display the result for the current column with the exact format requested
        print(f"The column {col} is missing {missing_percentage:.1f}% of data!")

# Call the function to display missing value percentages
count_nan()
```

# 2.4 Transforming the data>Lastly as part of the assignment we will be transforming and normalizing the data values. In Exercise 3c you saw how to perform this for one column, here we will recreate it but for all the columns that can benefit from it (which truth be told is all of them). The solutions is also a bit different so you cant fully copy paste. In short; the objective is to normalize each column in the dataframe. Hint; Check out the pandas docs for an idea of how to use apply at the df[column] part>

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("winequality-red.csv")

# Function to normalize the column values
def normalize_values(col_value, min_val, max_val):
    normalized_val = (col_value - min_val) / (max_val - min_val)
    return normalized_val

# Main function to normalize the entire dataset
def main():
    for column in df.columns:
        min_val = df[column].min()  # Get the minimum value of the column
        max_val = df[column].max()  # Get the maximum value of the column
        df[column] = df[column].apply(lambda x: normalize_values(x, min_val, max_val))  # Apply normalization

# Call the main function to normalize the dataset
main()

# Display the normalized dataset summary
print("Normalized dataset summary:")
display(df.describe())
```
